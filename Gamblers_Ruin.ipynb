```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

GAMBLER'S RUIN

```python
class Gambler:
    def __init__(self, start_money=50, goal=100, betsize=1, winprob=0.5):
        self.start_money = start_money
        self.goal = goal
        self.betsize = betsize
        self.winprob = winprob
        self.lossprob = 1 - winprob
    
    def simulate(self):
        curr_money = self.start_money
        event = []
        while 0 < curr_money < self.goal :
            outcome = np.random.choice([self.betsize, -self.betsize], p=[self.winprob, self.lossprob])
            curr_money += outcome
            event.append(curr_money)
        return event

    
    def plot_simulations(self):
        plt.plot(self.simulate())

    def get_win_prob(self, n_simulations):
        wins, losses = 0, 0
        probs = []
        for sim in tqdm(range(1, n_simulations)):
            outcome = self.simulate()[-1]
            if outcome >= self.goal:
                wins += 1
            probs.append(wins / sim)

        return probs
```

Plot the results of 5 simulations

```python
g = Gambler(winprob=0.5)
g.plot_simulations()
g.plot_simulations()
g.plot_simulations()
g.plot_simulations()
g.plot_simulations()
```
